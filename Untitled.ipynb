```python
def func_batsman_avg(string):
    batsmanname1 = string
    teamname = df_deliveries.batting_team[df_deliveries.batsman == batsmanname1]
    teams = teamname.value_counts().reset_index()
    team_list = df_full.batting_team[~df_full.batting_team.isin(teams['index'])].unique()
    df_full_list = df_matches.merge(df_deliveries,left_on='id',right_on='match_id',how='left')
    
    for battingteam in teams['index']:
        for team in team_list:
            df_list1 = df_full_list[(df_full_list.batsman==batsmanname1) & (df_full_list.batting_team ==battingteam) & (df_full_list.winner ==battingteam) & (df_full_list.bowling_team == team)]
            df_list_count = df_full_list[(df_full_list.batsman==batsmanname1) & (df_full_list.batting_team ==battingteam) & (df_full_list.winner ==battingteam) & (df_full_list.bowling_team == team) & (df_full_list.player_dismissed == batsmanname1)]
            df_list2 = df_full_list[(df_full_list.batsman==batsmanname1) & (df_full_list.batting_team ==battingteam) & (df_full_list.winner !=battingteam) & (df_full_list.bowling_team == team)]
            df_list2_count = df_full_list[(df_full_list.batsman==batsmanname1) & (df_full_list.batting_team ==battingteam) & (df_full_list.winner !=battingteam) & (df_full_list.bowling_team == team) & (df_full_list.player_dismissed == batsmanname1)]
    
            avg = df_list1.batsman_runs.sum()/np.count_nonzero(np.unique(df_list_count.id))
            avg1 = df_list2.batsman_runs.sum()/np.count_nonzero(np.unique(df_list2_count.id))
        
            print(str(batsmanname1) + ' averages '+ str(round(avg,2)) + ' for ' + battingteam + ' against ' + team + ' in wins.')
            print(str(batsmanname1) + ' averages '+ str(round(avg1,2)) + ' for ' + battingteam + ' against ' + team + ' in losses.')
```

```python
def func_matchdetails(string1,string2):
    match_cols_list = ['id','season', 'city', 'date', 'team1', 'team2', 'toss_winner',
       'toss_decision', 'result', 'dl_applied', 'winner', 'win_by_runs',
       'win_by_wickets', 'player_of_match', 'venue',]
    df1 = df_matches[(df_matches.team1 == string1) & (df_matches.team2 == string2)]
    df2 = df_matches[(df_matches.team1 == string2) & (df_matches.team2 == string1)]
    df_match_dets = pd.concat([df1,df2])

    total_score = []
    total_wickets = []
    team1_score = []
    team1_wickets = []
    team2_score = []
    team2_wickets = []

    for gameid in df_match_dets.id:
        #print(gameid)
        df_score = df_deliveries[df_deliveries.match_id == gameid]
        total_score.append(df_score.total_runs.sum())
        df_team1_score = df_score[df_score.batting_team == string1]
        df_team2_score = df_score[df_score.batting_team == string2]
        
        team1_score.append(df_team1_score.total_runs.sum())
        team2_score.append(df_team2_score.total_runs.sum())
        #print(total_score)
        total_wickets.append(df_score.player_dismissed.count())
        team1_wickets.append(df_team1_score.player_dismissed.count())
        team2_wickets.append(df_team2_score.player_dismissed.count())
        #print(total_wickets)
    df_final = df_match_dets[match_cols_list]

    df_final['total_runs'] = total_score
    df_final['total_wickets'] = total_wickets
    
    df_final[string1 + ' total_runs'] = team1_score
    df_final[string1 + ' total_wickets'] = team1_wickets
    
    df_final[string2 + ' total_runs'] = team2_score
    df_final[string2 + ' total_wickets'] = team2_wickets
    
    return df_final
```

```python
func_matchdetails('Royal Challengers Bangalore','Mumbai Indians')
```

```python
def function_win_toss(string1):
    df_tosswin = df_matches[(df_matches.toss_winner == string1) & ((df_matches.team1 == string1) | (df_matches.team2 == string1))]
    df_tosslost = df_matches[(df_matches.toss_winner != string1) & ((df_matches.team1 == string1) | (df_matches.team2 == string1))]
    
    df_matcheswon = df_tosswin[df_tosswin.winner == string1]
    df_matcheslost = df_tosswin[df_tosswin.winner != string1]
    
    df_matcheswon1 = df_tosslost[df_tosslost.winner == string1]
    df_matcheslost1 = df_tosslost[df_tosslost.winner != string1]
    
    print(str(string1) + ' won ' + str(np.count_nonzero(df_matcheswon.id)) + ' games after winning the toss.')
    print(str(string1) + ' lost ' + str(np.count_nonzero(df_matcheslost.id)) + ' games after winning the toss.')
    print(str(string1) + ' won ' + str(np.count_nonzero(df_matcheswon1.id)) + ' games after losing the toss.')
    print(str(string1) + ' lost ' + str(np.count_nonzero(df_matcheslost1.id)) + ' games after losing the toss.')
```

```python
def func_avg_batting(string1,string2):
    df_batsmen_1 = df_deliveries[(df_deliveries.batsman == string1) & (df_deliveries.inning == 1)]
    df_batsmen_1_diss = df_deliveries[(df_deliveries.batsman == string1) & (df_deliveries.inning == 1) &(df_deliveries.player_dismissed ==string1)]
    df_batsmen_2 = df_deliveries[(df_deliveries.batsman == string1) & (df_deliveries.inning == 2)]
    df_batsmen_2_diss = df_deliveries[(df_deliveries.batsman == string1) & (df_deliveries.inning == 2) &(df_deliveries.player_dismissed ==string1)]
    
    df_batsmen_3 = df_deliveries[(df_deliveries.batsman == string1) & (df_deliveries.inning == 1) & (df_deliveries.bowling_team == string2)]
    df_batsmen_3_diss = df_deliveries[(df_deliveries.batsman == string1) & (df_deliveries.inning == 1) & (df_deliveries.bowling_team == string2) &(df_deliveries.player_dismissed ==string1)]
    df_batsmen_4 = df_deliveries[(df_deliveries.batsman == string1) & (df_deliveries.inning == 2) & (df_deliveries.bowling_team == string2)]
    df_batsmen_4_diss = df_deliveries[(df_deliveries.batsman == string1) & (df_deliveries.inning == 2) & (df_deliveries.bowling_team == string2) &(df_deliveries.player_dismissed ==string1)]
    
    avg1 = df_batsmen_1.batsman_runs.sum()/np.count_nonzero(np.unique(df_batsmen_1_diss.match_id))
    avg2 = df_batsmen_2.batsman_runs.sum()/np.count_nonzero(np.unique(df_batsmen_2_diss.match_id))
    avg3 = df_batsmen_3.batsman_runs.sum()/np.count_nonzero(np.unique(df_batsmen_3_diss.match_id))
    avg4 = df_batsmen_4.batsman_runs.sum()/np.count_nonzero(np.unique(df_batsmen_4_diss.match_id))
    
    strikerate1 = (df_batsmen_1.batsman_runs.sum()/np.count_nonzero(df_batsmen_1.match_id)) * 100
    strikerate2 = (df_batsmen_2.batsman_runs.sum()/np.count_nonzero(df_batsmen_2.match_id)) * 100
    strikerate3 = (df_batsmen_3.batsman_runs.sum()/np.count_nonzero(df_batsmen_3.match_id)) * 100
    strikerate4 = (df_batsmen_4.batsman_runs.sum()/np.count_nonzero(df_batsmen_4.match_id)) * 100
    
    #print(df_batsmen_1)
    #print(df_batsmen_1.batsman_runs.sum())
    #print(np.count_nonzero(np.unique(df_batsmen_1.match_id)))
    #print(np.count_nonzero(np.unique(df_batsmen_1_diss.match_id)))
    
    print(str(string1) + ' has an average of ' +str(round(avg1,2))+ ' and strike rate of ' +str(round(strikerate1,2))+ ' batting first overall.')
    print(str(string1) + ' has an average of ' +str(round(avg2,2))+ ' and strike rate of ' +str(round(strikerate2,2))+ ' batting second overall.')
    print(str(string1) + ' has an average of ' +str(round(avg3,2))+ ' and strike rate of ' +str(round(strikerate3,2))+ ' batting first against '+string2)
    print(str(string1) + ' has an average of ' +str(round(avg4,2))+ ' and strike rate of ' +str(round(strikerate4,2))+ ' batting second against '+string2)
    
```
